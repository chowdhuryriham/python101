```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel data into a Pandas dataframe
df = pd.read_csv('/content/sample_data/IHME-GBD_2019_DATA-7fcfdca9-1.csv')

# Group the data by year and cause_name, and calculate the sum of values
grouped_data = df.groupby(['year', 'cause_name'])['val'].sum().reset_index()

# Filter data for every 10 years
#grouped_data = grouped_data[grouped_data['year'] % 10 == 0]

# Pivot the data to create a new dataframe with cause_name as columns and year as index
pivoted_data = grouped_data.pivot(index='year', columns='cause_name', values='val')

# Extract the x and y data for the bar chart
x_data = pivoted_data.index
y_data = pivoted_data.values

# Define a list of colors for each cause_name
colors = ['#104E8B', '#1E90FF', '#00BFFF', '#68228B', '#BF3EFF']

# Create a new figure with a larger size
fig = plt.figure(figsize=(18, 10))

# Create a stacked bar chart using Matplotlib
plt.bar(x_data, y_data[:, 0], color=colors[0], label=pivoted_data.columns[0])
for i in range(1, y_data.shape[1]):
    plt.bar(x_data, y_data[:, i], bottom=y_data[:, :i].sum(axis=1), color=colors[i], label=pivoted_data.columns[i])

# to show all values in x axis
plt.xticks(x_data)

# Add labels, a legend, and a title to the chart
plt.xlabel('Year')
plt.ylabel('Death Per 100000')

# title
#plt.title('Example Stacked Bar Chart')

#to keep the legend outside
plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",mode="expand", borderaxespad=0, ncol=3)

# Display the chart
plt.show()
```
